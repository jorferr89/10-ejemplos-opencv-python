## Detección de Ojos con Haar Cascades

Se importan las librerías necesarias.

```python
import cv2
from google.colab.patches import cv2_imshow
```

Cargar el clasificador pre-entrenado

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

Cargar imagen

```python
image = cv2.imread('imagen-messi.jpg')
```

Convertir imagen en escala de grises

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

Detectar ojos en la imagen

```python
eyes = eye_cascade.detectMultiScale(gray_image, scaleFactor=2, minNeighbors=3)
```

Dibujar rectángulos alrededor de los ojos detectados y mostrar la imagen

```python
for (x, y, w, h) in eyes:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

cv2_imshow(image)
```
