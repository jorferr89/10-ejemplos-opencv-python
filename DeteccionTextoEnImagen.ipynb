## Detección de Texto en Imagen

Se importan las librerías y se descarga el archivo 'spa.traineddata'.

```python
import cv2
import pytesseract
from google.colab import files
import os
from matplotlib import pyplot as plt
```

Se descarga el archivo 'spa.traineddata' y se configura la variable de entorno TESSDATA_PREFIX

```python
!wget https://github.com/tesseract-ocr/tessdata/raw/main/spa.traineddata -P /usr/share/tesseract-ocr/4.00/tessdata/

pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
os.environ['TESSDATA_PREFIX'] = '/usr/share/tesseract-ocr/4.00/tessdata/'
```

Se carga la foto a color.

```python
image = cv2.imread('imagen-con-texto.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(image)
```

Se extrae el texto y se imprime. Muestra la cantidad de palabras extraídas.

```python
text = pytesseract.image_to_string(image, lang='spa')
print('Texto:', text)

count = len(text.split())
print('Cantidad de palabras:', count)
```

Adicional: se guarda el texto en un archivo '.txt' y se descarga.

```python
with open('text.txt', 'w') as file:
    file.write(text)

files.download('text.txt')
```
